## Transformación logarítmica simétrica que mantenga valores negativos
la transformación log-simétrica:

- Incluye valores negativos en el análisis.
- Hace más fácil visualizar y comparar las distribuciones de las variables.
- Mantiene la relación de magnitudes entre valores.

```python
def log_simetrico(x):
    if x >= 0:
        return np.log1p(x)  # Logaritmo natural clásico para valores positivos
    else:
        return -np.log1p(-x)  # Logaritmo simétrico para valores negativos

# Aplicar la transformación a las columnas relevantes
df['Log_Real'] = df['Real'].apply(log_simetrico)
df['Log_Previsto'] = df['Previsión'].apply(log_simetrico)
df['Log_Porcentaje'] = df['Porcentaje'].apply(log_simetrico)
```

```python
df[['Log_Real', 'Log_Previsto', 'Log_Porcentaje']].hist(bins=30, figsize=(10, 6))
plt.suptitle('Distribuciones transformadas (log-simétrico)')
plt.show()
```

```python
# Calcular el percentil 75% para cada variable transformada
p75_real = np.percentile(df['Log_Real'], 75)
p75_previsto = np.percentile(df['Log_Previsto'], 75)
p75_porcentaje = np.percentile(df['Log_Porcentaje'], 75)

# Filtrar los datos
df_filtered = df[
    (df['Log_Real'] <= p75_real) &
    (df['Log_Previsto'] <= p75_previsto) &
    (df['Log_Porcentaje'] <= p75_porcentaje)
]

# Visualizar las distribuciones filtradas
df_filtered[['Log_Real', 'Log_Previsto', 'Log_Porcentaje']].hist(bins=30, figsize=(10, 6))
plt.suptitle('Distribuciones filtradas (por debajo del percentil 75%)')
plt.show()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Antes del filtrado
df[['Log_Real', 'Log_Previsto', 'Log_Porcentaje']].hist(bins=30, ax=axes[0], color='blue')
axes[0][0].set_title('Log_Real (Original)')
axes[0][1].set_title('Log_Previsto (Original)')
axes[0][2].set_title('Log_Porcentaje (Original)')

# Después del filtrado
df_filtered[['Log_Real', 'Log_Previsto', 'Log_Porcentaje']].hist(bins=30, ax=axes[1], color='green')
axes[1][0].set_title('Log_Real (Filtrado)')
axes[1][1].set_title('Log_Previsto (Filtrado)')
axes[1][2].set_title('Log_Porcentaje (Filtrado)')

plt.suptitle('Comparación de Distribuciones: Antes y Después del Filtrado')
plt.tight_layout()
plt.show()
```

# Análisis de desviaciones

- El siguiente paso es calcular y analizar directamente las desviaciones entre lo previsto y lo ejecutado. Vamos a calcular dos métricas clave:

1. Diferencia Absoluta = Real - Previsión

2. Diferencia relativa (o porcentaje de desviación): Real - Previsión / Previsión * 100


```python
# Calcular las desviaciones
df['Diferencia_Absoluta'] = df['Real'] - df['Previsión']
df['Porcentaje_Desviación'] = ((df['Real'] - df['Previsión']) / df['Previsión']) * 100
```

```python
# Aplicar transformación logarítmica a las métricas (agregar un pequeño valor para evitar log(0))
df['Log_Diferencia_Absoluta'] = np.log1p(df['Diferencia_Absoluta'].abs())  # abs para manejar negativos
df['Log_Porcentaje_Desviación'] = np.log1p(df['Porcentaje_Desviación'].abs())

# Filtrar el rango de percentiles para valores representativos
p5_abs = np.percentile(df['Diferencia_Absoluta'], 5)
p95_abs = np.percentile(df['Diferencia_Absoluta'], 95)
p5_dev = np.percentile(df['Porcentaje_Desviación'], 5)
p95_dev = np.percentile(df['Porcentaje_Desviación'], 95)

df_filtrado = df[
    (df['Diferencia_Absoluta'] >= p5_abs) & (df['Diferencia_Absoluta'] <= p95_abs) &
    (df['Porcentaje_Desviación'] >= p5_dev) & (df['Porcentaje_Desviación'] <= p95_dev)
]

# Visualizar las distribuciones transformadas y filtradas
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Diferencia Absoluta - Transformada
df['Log_Diferencia_Absoluta'].hist(bins=30, ax=axes[0, 0], color='orange')
axes[0, 0].set_title('Log de Diferencia Absoluta (Transformada)')
axes[0, 0].set_xlabel('Log(Diferencia Absoluta)')
axes[0, 0].set_ylabel('Frecuencia')

# Porcentaje de Desviación - Transformado
df['Log_Porcentaje_Desviación'].hist(bins=30, ax=axes[0, 1], color='purple')
axes[0, 1].set_title('Log del Porcentaje de Desviación (Transformado)')
axes[0, 1].set_xlabel('Log(Porcentaje de Desviación)')
axes[0, 1].set_ylabel('Frecuencia')

# Diferencia Absoluta - Filtrada
df_filtrado['Diferencia_Absoluta'].hist(bins=30, ax=axes[1, 0], color='orange')
axes[1, 0].set_title('Diferencia Absoluta (Filtrada)')
axes[1, 0].set_xlabel('Diferencia Absoluta')
axes[1, 0].set_ylabel('Frecuencia')

# Porcentaje de Desviación - Filtrado
df_filtrado['Porcentaje_Desviación'].hist(bins=30, ax=axes[1, 1], color='purple')
axes[1, 1].set_title('Porcentaje de Desviación (Filtrado)')
axes[1, 1].set_xlabel('Porcentaje de Desviación')
axes[1, 1].set_ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

## Objetivo 3: Rendimiento por órgano y unidad gestora: Evaluar qué órganos o unidades gestoras son más eficientes en términos de alcanzar las metas de recaudación y cuáles presentan consistentemente una baja ejecución.

Objetivo:
Evaluar el porcentaje de ejecución (Porcentaje) para cada órgano y unidad gestora. Comparar el rendimiento promedio, la variabilidad, y los casos más extremos (bajos y altos).

**Pasos:**

- Agrupar los datos por:
                                    - Órgano de nombre.
                                    - Nombre de gestión de la unidad.
- Calcular las métricas clave:
                    
                    - Promedio del porcentaje de ejecución (Porcentaje).
                    - Desviación estándar (para evaluar la consistencia).
                    - Total recaudado (Real) y total previsto (Previsión) por órgano/unidad gestora.

```python
# Agrupación por órgano
rendimiento_organo = df.groupby('Órgano de nombre').agg({
    'Porcentaje': ['mean', 'std'],
    'Real': 'sum',
    'Previsión': 'sum'
}).reset_index()

# Renombrar columnas
rendimiento_organo.columns = ['Órgano de nombre', 'Porcentaje_Medio', 'Porcentaje_Desviación', 'Real_Total', 'Previsión_Total']

# Calcular ejecución total (Real vs Previsión)
rendimiento_organo['Ejecución_Total'] = (rendimiento_organo['Real_Total'] / rendimiento_organo['Previsión_Total']) * 100

# Agrupación por unidad gestora
rendimiento_unidad = df.groupby('Nombre de gestión de la unidad').agg({
    'Porcentaje': ['mean', 'std'],
    'Real': 'sum',
    'Previsión': 'sum'
}).reset_index()

# Renombrar columnas
rendimiento_unidad.columns = ['Nombre de gestión de la unidad', 'Porcentaje_Medio', 'Porcentaje_Desviación', 'Real_Total', 'Previsión_Total']

# Calcular ejecución total (Real vs Previsión)
rendimiento_unidad['Ejecución_Total'] = (rendimiento_unidad['Real_Total'] / rendimiento_unidad['Previsión_Total']) * 100
```
